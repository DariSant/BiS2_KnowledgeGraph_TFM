```python
import os
from google.colab import userdata

# --- 1. CONFIGURATION ---
USERNAME = "DariSant"
REPO = "BiS2_KnowledgeGraph_TFM"
EMAIL = "your_email@example.com"
NAME = "DariSant"

try:
    token = userdata.get('GITHUB_TOKEN')
except:
    print("‚ùå Error: GITHUB_TOKEN not found in Secrets.")

# --- 2. NUKE & RE-INIT ---
print("üßπ Cleaning up old Git history (removing secrets)...")
!rm -rf .git

print("‚ú® Initializing fresh Git repository...")
!git init
!git config --global user.email "$EMAIL"
!git config --global user.name "$NAME"

# --- 3. RESTORE .GITIGNORE ---
# We recreate this to ensure we don't accidentally add Google Drive trash files
gitignore_content = """
*.gdoc
*.gsheet
*.gslides
.ipynb_checkpoints
__pycache__
secrets.json
.env
"""
with open(".gitignore", "w") as f:
    f.write(gitignore_content)

# --- 4. PREPARE CLEAN COMMIT ---
print("‚ûï Adding clean files...")
!git add .
!git commit -m "Initial release: Cleaned pipeline and notebooks"

# --- 5. FORCE PUSH ---
auth_url = f"https://{USERNAME}:{token}@github.com/{USERNAME}/{REPO}.git"

print(f"üîó Connecting to {REPO}...")
!git remote add origin "$auth_url"

print("üöÄ Force Pushing clean version...")
!git push origin main --force
```
